Menggabungkan (merge) tabel day_df dan hour_df

```python
day_hour_df = pd.merge(
    left=day_df,
    right=hour_df,
    how="left",
    left_on="dteday",
    right_on="dteday"
)
day_hour_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana performa penyewaan rental sepeda selama sebulan?


```python
print(day_hour_df.columns)
```

```python
total_rentals = day_hour_df['cnt_x'].sum()
print("Total Rentals:", total_rentals)
```

```python
day_hour_df['date'] = pd.to_datetime(day_hour_df['dteday'], format='%d%m%Y')

monthly_rentals = day_hour_df.groupby(pd.Grouper(key='date', freq='M'))['cnt_x'].sum()

plt.figure(figsize=(10, 5))
plt.plot(monthly_rentals.index, monthly_rentals.values, marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh cuaca terhadap penyewaan rental sepeda?

```python
seasonal_rentals = day_hour_df.groupby('season_x')['cnt_x'].mean()
plt.figure(figsize=(10, 6))
plt.bar(seasonal_rentals.index, seasonal_rentals.values, color='skyblue')
plt.title('Average Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Average Rentals')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

- Dicoding


## Analisis Lanjutan (Opsional)

Daily Rentals


```python
plt.figure(figsize=(12, 6))
plt.plot(day_hour_df['dteday'], day_hour_df['cnt_x'])
plt.title('Daily Bike Rentals')
plt.xlabel('Date')
plt.ylabel('Number of Rentals')
plt.show()
```

Yearly Rentals

```python
yearly_rentals = day_hour_df.groupby(day_hour_df['dteday'].dt.year)['cnt_x'].sum()
plt.figure(figsize=(12, 6))
plt.bar(yearly_rentals.index, yearly_rentals.values)
plt.title('Yearly Bike Rentals')
plt.xlabel('Year')
plt.ylabel('Number of Rentals')
plt.show()
```

Weekday vs Weekend

```python
weekday_rentals = day_hour_df[day_hour_df['workingday_x'] == 1]['cnt_x'].mean()
weekend_rentals = day_hour_df[day_hour_df['workingday_x'] == 0]['cnt_x'].mean()

plt.figure(figsize=(6, 6))
plt.bar(['Weekday', 'Weekend'], [weekday_rentals, weekend_rentals])
plt.title('Weekday vs. Weekend Rentals')
plt.xlabel('Day')
plt.ylabel('Average Rentals')
plt.show()
```

Working Day vs. Non-Working Day

```python
workingday_rentals = day_hour_df[day_hour_df['workingday_x'] == 1]['cnt_x'].mean()
non_workingday_rentals = day_hour_df[day_hour_df['workingday_x'] == 0]['cnt_x'].mean()

plt.figure(figsize=(6, 6))
plt.bar(['Working Day', 'Non-Working Day'], [workingday_rentals, non_workingday_rentals])
plt.title('Working Day vs. Non-Working Day Rentals')
plt.xlabel('Day Type')
plt.ylabel('Average Rentals')
plt.show()
```

Casual vs. Registered Rentals

```python
plt.figure(figsize=(12, 6))
plt.plot(day_hour_df['dteday'], day_hour_df['casual_x'], label='Casual')
plt.plot(day_hour_df['dteday'], day_hour_df['registered_x'], label='Registered')
plt.title('Casual vs. Registered Rentals')
plt.xlabel('Date')
plt.ylabel('Number of Rentals')
plt.legend()
plt.show()
```
